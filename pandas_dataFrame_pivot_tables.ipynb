```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic
```

```python
titanic.groupby("sex")["survived"].describe()
#"sex"e göre grupla, "survived"=1(kurtuldu) olanları describe et
```

```python
titanic.groupby("sex")[["survived"]].describe()#dataframe olsun istiyosak
#iki tane köşeli parantez koyabiliriz.
```

```python
titanic.groupby(["sex", "class"])[["survived"]].describe()
#hem "sex"e hem "class"a göre grupla ve üsttekinin aynısını yap
```

```python
titanic.groupby(["sex", "class"])[["survived"]].aggregate("mean")
#hem "sex"e hem "class"a göre grupla ve hayatta kalanların ortalamasını göster
```

```python
titanic.groupby(["sex", "class"])[["survived"]].aggregate("mean").unstack()
#üsttekinin aynısı ancak farklı okuma olanağı
```

```python
#pivot ile table
```

```python
titanic.pivot_table("survived", index = "sex", columns = "class")
#üsttekilerin hepsini daha az çaba ve kodla yaptık
```

```python
titanic.age #bu age'i kategorik bir değişken yapmak için aşağıdakileri uygulayacağız
```
